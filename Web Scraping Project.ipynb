```python
import requests
from bs4 import BeautifulSoup
import json
import re
import markdown

# Function to extract content from a blog page
def extract_content(blog_url):
    response = requests.get(blog_url)
    if response.status_code != 200:
        return None

    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.find('h1', class_='art_title').text.strip()
    
    # Extracting content and converting tables to markdown
    content = soup.find('div', class_='art_text')
    tables = content.find_all('table')
    for table in tables:
        markdown_table = ""
        rows = table.find_all('tr')
        for row in rows:
            cells = row.find_all(['th', 'td'])
            markdown_table += "|".join(cell.get_text().strip() for cell in cells) + "\n"
        table.replace_with(markdown_table)

    content = content.get_text(separator="\n").strip()
    return {"title": title, "content": content}

# Function to scrape the provided URL
def scrape_website(url):
    data = []
    page_number = 1

    while True:
        paginated_url = f"{url}?page={page_number}"
        response = requests.get(paginated_url)

        if response.status_code != 200:
            break  # Exit if the page doesn't exist or other issue occurs

        soup = BeautifulSoup(response.content, 'html.parser')
        articles = soup.find_all('div', class_='article')

        if not articles:
            break  # No more articles to scrape

        for article in articles:
            blog_url = article.find('a')['href']
            content = extract_content(blog_url)
            if content:
                data.append(content)

        page_number += 1

    return data

# Scrape the website and save the data to a JSON file
url_to_scrape = "https://www.legalserviceindia.com/articles/articles.html"
data = scrape_website(url_to_scrape)

with open('scraped_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)

print("Scraping completed and data saved to scraped_data.json.")
```
